```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

from tensorflow import keras
import tensorflow as tf
from keras.layers import *
from keras import backend as K
```

```python
df = pd.read_csv('../Resultados/df_train.csv')
```

```python
y = df['rating_num']
X = df.drop(['rating_num', 'row_id'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
```

```python
def coeff_determination(y_true, y_pred):
    SS_res =  K.sum(K.square( y_true-y_pred )) 
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) ) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
model = keras.Sequential()
model.add(Dense(64, input_dim=450, activation='tanh')) #ok 64 - bom, 32 -ruim, 128-quase bom 100 ruim
# model.add(Dropout(.2))
model.add(Dense(20, activation='relu')) #32 ->2.10, 64 ->2.9 20->1.8, 10-.3.6 tanh uma merda
# model.add(Dropout(.2))
# model.add(Dense(16, activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
opt = keras.optimizers.Nadam(learning_rate=0.005) #0.001 e 0.01-pior 0.005 - melhor 0.005 - ok
model.compile(optimizer= opt, loss='mse', metrics=['mse'])#coeff_determination
history = model.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
pred_val= model.predict(X_test)
r2_score(pred_val, y_test)
```

```python
test = pd.read_csv('../Resultados/df_test.csv')
row_id = test['row_id']
test.drop(['row_id'], axis=1, inplace=True)
test
```

```python
pred = model.predict(test)
pred.shape
```

```python
submission = pd.DataFrame()
submission['row_id'] = row_id
submission['rating_num'] = pred
submission
```
